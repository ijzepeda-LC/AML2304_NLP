### Word Lengthening:
a word might onoly have same letter repeated twice: book, cool, collapse

```python
import re
def remove_lenghtening(text):
    pat= re.compile(r'(.)\1{2,}')
    return pat.sub(r'\1\1', text)
print(remove_lenghtening("booook"))
print(remove_lenghtening("booookk"))
print(remove_lenghtening("booookkkk"))
print(remove_lenghtening("booookkokookkk"))


```

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
```

```python
analyzer = SentimentIntensityAnalyzer()
```

```python
sentence = "The car is cool"
analyzer.polarity_scores(sentence)   
```

compund is the intensity of the value, extreme negative -1, extreme positive  
The compund score is a metric that calculates the sum of all the lexicon ratings   
which have been noramlized between -1(most extreme negative) and +1 (extreme positive)

### Vader Key points
- The exclamation will increase the positivity
- capitals on a word will emphasize it
- all capitals is SCREAMING so it is negative
- using a modifier like "bit", will decrease the intension



```python
def test_text(_text):
    print(analyzer.polarity_scores(_text))

test_text("The car is cool")  
test_text("The car is cool!")  
test_text("The car is cool!!!!!")  
test_text("The car is COOL")  
test_text("THE CAR IS COOL")  



```

### DEgree modifiers

```python
test_text ("The car is a bit cool.")
test_text ("The car is a bit cool")
```

### Conjunction

```python
test_text("I enjoyed your hotel food")
test_text("I enjoyed your hotel food, but I didn't like the way waiters offer their services")
test_text("I enjoyed your hotel food, but I did like more the way waiters offer their services")
test_text("I enjoyed your hotel food, I did like more the way waiters offer their services")
test_text("I enjoyed your hotel food, and I did like more the way waiters offer their services")
```

```python
test_text("your hotel service is great!")
test_text("your hotel service is not that great!")

```
