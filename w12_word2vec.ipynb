```python
import gensim
from nltk import sent_tokenize
from nltk import word_tokenize
```

```python
text = "This is or first sentence. This is the second one.  And one more."

sentences = sent_tokenize(text)
tokens = [word_tokenize(sentence) for sentence in sentences]
```

```python
from gensim.models import Word2Vec
# model = Word2Vec(tokens, min_count=1, size=50) # size is dimension, now is vector_size
model = Word2Vec(tokens, min_count=1, vector_size=50) 
# The more size value, will require more time to train, but will be more accurate
```

```python
print(model) #alpha is the learning rate
```

```python
# words=list(model.wv.vocab) #it changed on gensim4
words = model.wv.index_to_key
print(words)
```

**The vector size, is the number of vector_size;** The computation cost will increase but also increase the accuracy

```python
# print(model.wv['sentence'])
for word in words:
    print(f"{word} (size = {(len(model.wv[word]))}):")
    print(model.wv[word])
```
