```python
import pandas as pd
from gensim.models import Word2Vec
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

# Sample data
data = {
    'Disease_Name': ['Atrial fibrillation', 'Abdominal aortic aneurysm', 'Hyperhidrosis', 'Absence seizure', 'Achalasia'],
    'combined_text': ['In a typical heart...', 'An abdominal aortic aneurysm occurs...', 'Hyperhidrosis...', 'Absence seizures involve...', 'Achalasia is a rare disorder...']
}
df = pd.DataFrame(data)

# Preprocess data: Simple tokenization and lowercase conversion
df['tokens'] = df['combined_text'].str.lower().str.split()

# Train Word2Vec or load a pre-trained model
model = Word2Vec(sentences=df['tokens'], vector_size=100, window=5, min_count=1, workers=4)
model.train(df['tokens'], total_examples=model.corpus_count, epochs=10)

# Create a function to average word vectors for a text
def average_word_vectors(words, model, vocabulary, num_features):
    feature_vector = np.zeros((num_features,),dtype="float64")
    nwords = 0.
    
    for word in words:
        if word in vocabulary: 
            nwords = nwords + 1.
            feature_vector = np.add(feature_vector, model.wv[word])
    
    if nwords:
        feature_vector = np.divide(feature_vector, nwords)
        
    return feature_vector

vocab = set(model.wv.index_to_key)
df['avg_vector'] = df['tokens'].apply(lambda x: average_word_vectors(x, model, vocab, 100))

def predict_disease(user_input):
    user_input = user_input.lower().split()
    user_vector = average_word_vectors(user_input, model, vocab, 100)
    
    similarities = cosine_similarity([user_vector], df['avg_vector'].tolist())
    related_disease_idx = similarities.argsort()[0][-1]
    
    return df.iloc[related_disease_idx]['Disease_Name']

user_description = "I feel a fast and pounding heart rate..."
predicted_disease = predict_disease(user_description)
print(predicted_disease)
```

```python
import pandas as pd
import spacy
from spacy.lang.en.stop_words import STOP_WORDS
import string
from sklearn.metrics.pairwise import cosine_similarity

# Load spaCy model with pre-trained embeddings
nlp = spacy.load("en_core_web_lg")

# Preprocessing function
def preprocess(text):
    text = text.lower()
    text = text.translate(str.maketrans('', '', string.punctuation))
    doc = nlp(text)
    tokens = [token.lemma_ for token in doc if token.lemma_ != '-PRON-' and token.lemma_ not in STOP_WORDS]
    return ' '.join(tokens)


df = df_raw.copy()
df=df.rename({'Diesase_Name':'Disease_Name'},axis=1)
del df['Unnamed: 0']
```

```python
df['combined_text'] = df['overview'] + ' ' + df['symptoms_elements'] + ' ' + df['symptoms_paragraph']
df['combined_text'] = df['combined_text'].astype(str)

# Preprocess the data
df['cleaned_text'] = df['combined_text'].apply(preprocess)

# Convert the cleaned text into vector representations using spaCy
df['vectors'] = df['cleaned_text'].apply(lambda x: nlp(x).vector)

def predict_disease(user_input):
    user_input_cleaned = preprocess(user_input)
    user_vector = nlp(user_input_cleaned).vector
    
    similarities = [cosine_similarity([user_vector], [vec]) for vec in df['vectors']]
    related_disease_idx = similarities.index(max(similarities))
    
    return df.iloc[related_disease_idx]#['Disease_Name']
```

```python
user_description = "I feel a sudden stop and loss in activity..."
predicted_disease = predict_disease(user_description)
print(predicted_disease)
```
